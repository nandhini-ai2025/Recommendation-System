```python
#1.
import pandas as pd
from sklearn.feature_selection import mutual_info_classif

df = pd.DataFrame({
    'Age': [25, 32, 47, 51, 62],
    'Salary': [50000, 52000, 60000, 58000, 62000],
    'Purchased': [0, 1, 1, 0, 1]
})

X = df[['Age', 'Salary']]
y = df['Purchased']

mi_scores = mutual_info_classif(X, y)
mi_scores
```

```python
#2.
import pandas as pd

data = pd.DataFrame({
    'User': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Item': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Orange'],
    'Rating': [5, 3, 4, 2, 5, 4]
})

# Item popularity (sum of ratings)
popularity = data.groupby('Item')['Rating'].sum()

# Recommend top 2 items
top_items = popularity.sort_values(ascending=False).head(2)
top_items
```

```python
#3.
from sklearn.datasets import load_diabetes
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import pandas as pd

# Load dataset
X, y = load_diabetes(return_X_y=True, as_frame=True)

# Define model
model = LinearRegression()

# Apply RFE to select top 5 features
rfe = RFE(estimator=model, n_features_to_select=5)
rfe.fit(X, y)

# Get selected feature names
selected_features = X.columns[rfe.support_]

print("Selected features:", list(selected_features))
```

```python
#4.
from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectKBest, mutual_info_classif
import pandas as pd

# Load dataset
X, y = load_digits(return_X_y=True, as_frame=True)

# Rename columns to pixel-style names (for expected output format)
X.columns = [f"pixel_{i//8}_{i%8}" for i in range(X.shape[1])]

# Apply SelectKBest with mutual information
selector = SelectKBest(score_func=mutual_info_classif, k=8)
X_new = selector.fit_transform(X, y)

# Get selected feature names
selected_features = X.columns[selector.get_support()]

print("Top 8 features using mutual info:", list(selected_features))
```

```python
#5.
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
import pandas as pd

# Load dataset
X, y = load_breast_cancer(return_X_y=True, as_frame=True)

# Train Logistic Regression model
log_reg = LogisticRegression(max_iter=5000, solver='liblinear')
log_reg.fit(X, y)

# Feature selection using SelectFromModel
selector = SelectFromModel(estimator=log_reg, prefit=True)
X_selected = selector.transform(X)

# Get selected feature names
selected_features = X.columns[selector.get_support()]

print("Selected features:", list(selected_features))
```
